# 1. Problem Statement

The Avocado dataset has two types as  Organic & Conventional Type .We need to predict the Average price using Regression model from year 2013,2014 ,2015, 2016, 2017 and 2018 data.

It has following features in the Data set 

    Date - The date of the observation
    
    AveragePrice - the average price of a single avocado
    
    type - conventional or organic
    
    year - the year
    
    Region - the city or region of the observation
    
    Total Volume - Total number of avocados sold
    
    4046 - Total number of avocados with PLU 4046 sold
    
    4225 - Total number of avocados with PLU 4225 sold
    
    4770 - Total number of avocados with PLU 4770 sold
    
    Here there needs to be two Output features .
    
    1.AveragePrice
    2.Region

# Applying Linear Regression approach

```python
features = dfavocado3.drop("AveragePrice",axis= 1)
target = dfavocado3["Total Volume"]
```

```python
from sklearn.preprocessing import MinMaxScaler
mns =   MinMaxScaler()
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Logistic Regression model
for i in range(0,100):
        features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=i)
        lr.fit(features_train,target_train)
        pred_train = lr.predict(features_train)
        pred_test = lr.predict(features_test)
        print(f"At random state{i}  :- {r2_score(target_train,pred_train)}")
        print(f"At random state{i} : {r2_score(target_test,pred_test)}")
        print("\n")
```
